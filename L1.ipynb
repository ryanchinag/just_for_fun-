### Python對於「縮排」是非常嚴格的，如果沒有縮排或沒有對齊，程式在執行時立刻會出現錯誤訊息

### 將電腦程式轉成機器語言的方法有兩類: 

1. 編譯: 是將某一由電腦語言所撰寫的程式，經過編譯器，將整支程式轉換成機器語言，如果編譯的過程沒有錯誤，這支機器語言的程式就可以隨時執行。

2. 直譯: 這種電腦語言所寫的程式在執行時，經過直譯器 (interpreter) 的轉換，每執行一行就將一行轉換機器語言，一行一行經過直譯→執行→直譯→執行這樣的順序，反複執行到程式的最後一行。

### Relational Operators 關係運算子
用於比較兩個運算元的大小是否相等，若比較的結果為「真」，就傳回True，若結果為「假」，就傳回False。
* `==` if the values of two operands are equal, then the condition becomes true  結果為True時, 左式等於右式
* `!=` if values of two operands are not equal, then condition becomes true  結果為True時, 左式不等於右式
* `>`  結果為True時, 左式大於右式
* `>=` 結果為True時, 左式大於或等於右式
* `<`  結果為True時, 左式小於右式
* `<=` 結果為True時, 左式小於或等於右式

#### Example

```python
1==2
```

```python
x!=y
```
